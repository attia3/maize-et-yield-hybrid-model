```python
maize_df = pd.concat([ISMA,GIZA,SKHA,ASWAN],ignore_index=True)
maize_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
maize_df['SiteName'] = enc.fit_transform(maize_df['SiteName'])
maize_df['Texture'] = enc.fit_transform(maize_df['Texture'])
maize_df['OC'] = enc.fit_transform(maize_df['OC'])
maize_df['AWC'] = enc.fit_transform(maize_df['AWC'])
#maize_df['Yr'] = enc.fit_transform(maize_df['Yr'])
maize_df['DOY'] = enc.fit_transform(maize_df['DOY'])
maize_df['Irri'] = enc.fit_transform(maize_df['Irri'])
maize_df['Cultivar'] = enc.fit_transform(maize_df['Cultivar'])
maize_df['Compost'] = enc.fit_transform(maize_df['Compost'])
maize_df['CO2'] = enc.fit_transform(maize_df['CO2'])
print(np.unique(maize_df['Yr']))
maize_df.head()
```

```python
trainging_data = maize_df[maize_df['Yr'] < 2017]
testing_data = maize_df[maize_df['Yr'] >= 2017]

train_y = trainging_data['GY']
train_x = trainging_data.drop(['GY','Trt','SiteName','ET','CO2e'], axis=1,inplace=False)

test_y = testing_data['GY']
test_x = testing_data.drop(['GY','Trt','SiteName','ET','CO2e'], axis = 1, inplace=False)

scaler = MinMaxScaler()
train_x_scaled = scaler.fit_transform(train_x)
test_x_scaled = scaler.transform(test_x)
```

```python
from sklearn.ensemble import RandomForestRegressor
base_model = RandomForestRegressor(n_estimators = 10, random_state = 42)

base_model.fit(train_x_scaled,train_y)
ypred = base_model.predict(test_x_scaled)
```

```python
from sklearn import metrics 
print("MSE:",metrics.mean_squared_error(test_y,ypred)) 
print("MAE:",metrics.mean_absolute_error(test_y,ypred)) 
print("RMSE:",np.sqrt(metrics.mean_squared_error(test_y,ypred))) 
print("r2_score:",metrics.r2_score(test_y,ypred))
print('rrmse:',np.sqrt(np.mean( (ypred-test_y)**2)) / (max(test_y)-min(test_y))*100)
```

```python
from hyperopt import STATUS_OK, Trials, fmin, hp, tpe
```

```python
space = {
#    'bootstrap': hp.choice('bootstrap', ['True','Fasle']),
#    'criterion': hp.choice('criterion',['squared_error', 'absolute_error', 'poisson']),
    'max_depth': hp.uniform('max_depth', 5, 20), 
#    'max_features': hp.choice('max_features',[None ,'sqrt','log2']),
    'min_samples_leaf': hp.uniform('min_samples_leaf',   1, 5),
    'min_samples_split': hp.uniform('min_samples_split', 2, 6),
    'n_estimators': hp.uniform('n_estimators', 100, 500)
}
```

```python
def objective(space):
    rf=RandomForestRegressor(
                    n_estimators = int(space['n_estimators']), max_depth = int(space['max_depth']),
                    min_samples_leaf = int(space['min_samples_leaf']),min_samples_split=int(space['min_samples_split']))
    
    rf.fit(train_x_scaled, train_y)
    

    #pred = xgbr.predict(test_x_scaled)
    cv = RepeatedKFold(n_splits=3, n_repeats=2, random_state = 1)
    scores = cross_val_score(rf, train_x_scaled, train_y, scoring= 'neg_root_mean_squared_error', cv=cv, n_jobs=-1)
    scores = np.absolute(scores)
    print("SCORE:", scores.mean())
    return {'loss': scores.mean(), 'status': STATUS_OK }
```

```python
trials = Trials()

best_hyperparams = fmin(fn = objective,
                        space = space,
                        algo = tpe.suggest,
                        max_evals = 20,
                        trials = trials)
```
